```python
# Importando as bibliotecas necessárias
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Carregar os dados
import pandas as pd
data_set = pd.read_csv('https://raw.githubusercontent.com/celsocrivelaro/simple-datasets/main/user_data.csv')

# Definindo os atributos e o rótulo
x = data_set.iloc[:, [2, 3]].values
y = data_set.iloc[:, 4].values

# Dividindo o conjunto de dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# -----------------------
# Redes Neurais (MLPClassifier)
# -----------------------
mlp_clf = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42)
mlp_clf.fit(x_train, y_train)

# Previsão e avaliação
y_pred_mlp = mlp_clf.predict(x_test)
print("Redes Neurais - Accuracy:", accuracy_score(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))

# -----------------------
# SVM (Support Vector Machine)
# -----------------------
svm_clf = SVC(kernel='linear', random_state=42)
svm_clf.fit(x_train, y_train)

# Previsão e avaliação
y_pred_svm = svm_clf.predict(x_test)
print("SVM - Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))

# Exemplos de previsão usando os modelos
sample_input = [[30, 200000]]
print("Predição MLP:", mlp_clf.predict(sample_input))
print("Predição SVM:", svm_clf.predict(sample_input))
```
